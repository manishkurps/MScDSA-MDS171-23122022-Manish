Create a Python Class for managing student details and marks.
Define the class and implement the methods of the student class in a menu-driven program for different types of users

```python
class Student:
    def __init__(self, Registerno, name):
        self.Registerno = Registerno
        self.name = name
        self.marks = {}

    def add_mark(self, subject, subjectmark):
        self.marks[subject] = subjectmark

    def display_details(self):
        print(f"Student Register No : {self.Registerno}")
        print(f"Name: {self.name}")
        print("Marks:")
        for subject, subjectmark in self.marks.items():
            print(f"{subject}: {subjectmark}")
    

# Menu-driven program
students = {}  # Store student objects

while True:
    print("\n Enter a choice :")
    print("1. Add a student")
    print("2. Add marks for a student")
    print("3. Display student details")
    print("4. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        Registerno = input("Enter student Register No : ")
        name = input("Enter student name: ")
        student = Student(Registerno, name)
        students[Registerno] = student
        print("Student added successfully!")

    elif choice == "2":
        Registerno = input("Enter student Register No : ")
        if Registerno in students:
            subject = input("Enter subject: ")
            subjectmark = float(input("Enter subjectmark: "))
            students[Registerno].add_mark(subject, subjectmark)
            print("Mark added successfully!")
        else:
            print("Student not found.")

    elif choice == "3":
        Registerno = input("Enter Register no. : ")
        if Registerno in students:
            students[Registerno].display_details()
        else:
            print("Student not found.")

    elif choice == "4":
        print("Exiting the program.")
        break

    else:
        print("Invalid choice. Please select a valid option.")
```
