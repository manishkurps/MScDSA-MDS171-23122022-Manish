```python
# Function to input sales data from the user
def input_sales_data():
    sales_data = []
    while True:
        product = input("Enter the product name (or 'done' to finish): ")
        if product.lower() == 'done':
            break
        revenue = float(input("Enter the revenue: "))
        cost = float(input("Enter the cost: "))
        sale = {'Product': product, 'Revenue': revenue, 'Cost': cost}
        sales_data.append(sale)
    return sales_data

# Function to calculate total revenue without using sum
def calculate_total_revenue(data):
    total_revenue = 0
    for sale in data:
        total_revenue += sale['Revenue']
    return total_revenue

# Function to calculate total profit without using sum
def calculate_total_profit(data):
    total_profit = 0
    for sale in data:
        total_profit += sale['Revenue'] - sale['Cost']
    return total_profit

# Function to calculate average revenue and profit per product
def calculate_average_metrics(data):
    num_products = len(data)
    if num_products == 0:
        return 0, 0
    
    total_revenue = calculate_total_revenue(data)
    total_profit = calculate_total_profit(data)
    
    average_revenue = total_revenue / num_products
    average_profit = total_profit / num_products
    
    return average_revenue, average_profit

# Function to find the best-selling product without using max
def find_best_selling_product(data):
    product_counts = {}
    best_selling_product = None
    max_count = 0

    for sale in data:
        product = sale['Product']
        if product in product_counts:
            product_counts[product] += 1
        else:
            product_counts[product] = 1
        
        if product_counts[product] > max_count:
            max_count = product_counts[product]
            best_selling_product = product
    
    return best_selling_product

# Input sales data from the user
print("Enter 'done' in product name when finished.")
sales_data = input_sales_data()

# Display the product selected with revenue and cost
print("\nProduct list with their revenue and cost:\n")
for sale in sales_data:
    print(f"Product: {sale['Product']}, Revenue: {sale['Revenue']}, Cost: {sale['Cost']}")

# Calculate and display the total revenue and profit
total_revenue = calculate_total_revenue(sales_data)
total_profit = calculate_total_profit(sales_data)
print("\nE-commerce Sales Analysis:")
print("Total Revenue: ", total_revenue)
print("Total Profit: ", total_profit)

# Calculate and display the average revenue and profit per product
average_revenue, average_profit = calculate_average_metrics(sales_data)
print("Average Revenue per Product: ", average_revenue)
print("Average Profit per Product: ", average_profit)

# Display the best-selling product
best_selling_product = find_best_selling_product(sales_data)
print("Best-Selling Product:", best_selling_product)
```
